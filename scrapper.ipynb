# Scraping

```python
import numpy as np
import pandas as pd
import bs4
import requests
from time import sleep
from random import randint
```

```python
#declaring lists of results
# Living Wage Table
df_wages = []

# Expense Table
df_expenses =[]

# Salary Table
states = []
occupational_area = []
typical_salary = []


#array of page numbers
pages = np.arange(1,57,1)
```

```python
#Looping through each state page
for page in pages:
    #site doesn't have a a few pages
    if page not in (3,7,14,43,52):
        source = requests.get('https://livingwage.mit.edu/states/'+str(page))
        soup = bs4.BeautifulSoup(source.text,'lxml')
        sleep(randint(2,5))

        block = soup.find('table',class_='results_table table-striped salary_table').tbody.find_all('tr')

        #Salaries Table
        for occ in block:
            state = soup.find('div',class_ ='container').h1.text
            states.append(state)

            occupation = occ.find('td',class_='text').text.strip()
            occupational_area.append(occupation)

            salary = occ.find('td',class_={'black','red'}).text.strip()
            typical_salary.append(salary)
        
        #Living Wages Table
        dfs = pd.read_html('https://livingwage.mit.edu/states/'+str(page))
        living_wage_table = dfs[0]
        living_wage_table['State']= soup.find('div',class_ ='container').h1.text
        df_wages.append(living_wage_table)

        #Expenses Table
        expenses_table = dfs[1]
        expenses_table['State']= soup.find('div',class_ ='container').h1.text
        df_expenses.append(expenses_table)
    
    else:
        pass
```

```python
df_expenses = pd.concat(df_expenses)
df_expenses
```

```python
df_wages = pd.concat(df_wages)
df_wages
```

```python
df_salaries = pd.DataFrame(zip(states,occupational_area,typical_salary),columns=['State','Occupational Area', 'Typical Salary'])
df_salaries
```

# Data Cleaning
## Salary Table

```python
# Fuction to extract state name
def extract_state(x):
    if 'New' in x.split(' ') or 'Rhode' in x.split(' ') or 'West' in x.split(' ')or 'North' in x.split(' ') or 'South'in x.split(' '):
        return ' '.join(x.split(' ')[-2:])
    elif 'District' in x.split(' '):
        return ' '.join(x.split(' ')[-3:])
    else:
         return ' '.join(x.split(' ')[-1:])
```

```python
# For Salary table, only remove the $ sign and rename column to reflect this.
df_salaries['Typical Salary'] = df_salaries['Typical Salary'].str.replace('$','')
df_salaries.rename(columns={'Typical Salary':'Typical Salary in $'}, inplace=True)
df_salaries['State'] = df_salaries['State'].map(extract_state)
df_salaries
```

## Wages Table

```python
#Flatten the sub columns (multi-index) and remove the $
df_wages_clean = df_wages.copy()
```

```python

df_wages_clean.columns = df_wages_clean.columns.map(' & '.join)

# Rename columns, remove $ and rearange columns (State first)

for col in df_wages_clean.columns:
    df_wages_clean[col] = df_wages_clean[col].map(lambda x: x.strip('$'))
    df_wages_clean.rename(columns={col:'Wages for '+col+' ($)'},inplace=True)

df_wages_clean.rename(columns={'Wages for Unnamed: 0_level_0 & Unnamed: 0_level_1 ($)':'Wage Type','Wages for State &  ($)':'State'},inplace=True)

cols = df_wages_clean.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_wages_clean = df_wages_clean[cols]
df_wages_clean['State'] = df_wages_clean['State'].map(extract_state)

# Melting the different number of wages
df_wages_clean = df_wages_clean.melt(id_vars=['State', 'Wage Type'], value_vars=df_wages_clean.columns[2:])
df_wages_clean.rename(columns={'variable':'Household', 'value':'Wage ($)'}, inplace=True)
df_wages_clean['Household'] = df_wages_clean['Household'].apply(lambda x: x.split('for')[1].strip('($)'))
```

```python
df_wages_clean
```

## Expenses Table

```python
#Flatten the sub columns (multi-index)
df_expenses_clean = df_expenses.copy()

df_expenses_clean.columns = df_expenses_clean.columns.map(' & '.join)
df_expenses_clean

# Rename columns, remove $ and rearange columns (State first)

for col in df_expenses_clean.columns:
    df_expenses_clean[col] = df_expenses_clean[col].map(lambda x: x.strip('$'))
    df_expenses_clean.rename(columns={col:'Expenses for '+col+' ($)'},inplace=True)

df_expenses_clean.rename(columns={'Expenses for Unnamed: 0_level_0 & Unnamed: 0_level_1 ($)':'Expense Type','Expenses for State &  ($)':'State'},inplace=True)

cols = df_expenses_clean.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_expenses_clean = df_expenses_clean[cols]
df_expenses_clean['State'] = df_expenses_clean['State'].map(extract_state)

# Melting the different number of wages
df_expenses_clean = df_expenses_clean.melt(id_vars=['State', 'Expense Type'], value_vars=df_expenses_clean.columns[2:])
df_expenses_clean.rename(columns={'variable':'Household', 'value':'Expense ($)'}, inplace=True)
df_expenses_clean['Household'] = df_expenses_clean['Household'].apply(lambda x: x.split('for')[1].strip('($)'))

# Grab annual income before and after taxes and clean
df_required = df_expenses_clean[df_expenses_clean['Expense Type'].str.contains('annual')]
df_required['Expense Type'] = df_required['Expense Type'].apply(lambda x: 'Income after taxes' if 'after' in x else 'Income before taxes')
df_required.rename(columns={'Expense Type':'Income wrt Taxes','Expense ($)':'Income Required ($)'})

# Removing these from the expenses table.
df_expenses_clean = df_expenses_clean[~df_expenses_clean['Expense Type'].str.contains('annual')]
```

# Exports
